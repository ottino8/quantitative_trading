```r
##https://www.quantstart.com/articles/White-Noise-and-Random-Walks-in-Time-Series-Analysis/
set.seed(1)
acf(rnorm(1000))
```

```r
set.seed(1)
var(rnorm(1000, mean=0, sd=1))
```

```r
set.seed(4)
x <- w <- rnorm(1000)
for (t in 2:1000) x[t] <- x[t-1] + w[t]
plot(x, type="l")
```

```r
acf(x)
```

```r
acf(diff(x))
```
